### Merge Acquisition and Performance Data by Loan ID

```python
# Merge Acquisition & Performance Data for Q1 2015

merged_Q1_2015 = pd.merge(perf_q1_2015_subset, acq_q1_2015_subset, how='left', on='Loan_ID')
```

```python
merged_Q1_2015.drop(columns=['Servicer','Zero_Bal_Code','MSA','Rem_Months_to_Leg_Maturity','Adj_Months_to_Maturity','Curr_Act_UPB'],inplace=True)
```

```python
# Transform date variable
merged_Q1_2015['Date'] = pd.to_datetime(merged_Q1_2015['Date'])
merged_Q1_2015['Date'] = merged_Q1_2015['Date'].dt.date

merged_Q1_2015['Orig_Date'] = pd.to_datetime(merged_Q1_2015['Orig_Date'])
merged_Q1_2015['Orig_Date'] = merged_Q1_2015['Orig_Date'].dt.date
```

```python
# Sort by date and loan id
merged_Q1_2015.sort_values(by=['Loan_ID','Date'],inplace=True)
```

```python
merged_Q1_2015.info(verbose=True, null_counts=True)
```

```python
# Filter to loans originated in Q1 2015

merged_orig_Q1_2015 = merged_Q1_2015[(merged_Q1_2015['Orig_Date']>=datetime.date(2015,1,1)) & (merged_Q1_2015['Orig_Date']<=datetime.date(2015,3,1))]
```

```python
# Fill Mortgage Insurance NaN values with 0 (blank values = 'Uninsured')

values = {'Mort_Ins_Type': 0}
merged_orig_Q1_2015_nonan = merged_orig_Q1_2015.fillna(value=values)
```

```python
merged_orig_Q1_2015_nonan.info(verbose=True,null_counts=True)
```

```python
# Drop unknown loan statuses

merged_orig_Q1_2015_nonan_stat_fix = merged_orig_Q1_2015_nonan[merged_orig_Q1_2015_nonan['Curr_Loan_Status']!='X']
```

```python
# Transform loan status values from str to numeric

merged_orig_Q1_2015_nonan_stat_fix['Curr_Loan_Status'] = pd.to_numeric(merged_orig_Q1_2015_nonan_stat_fix['Curr_Loan_Status'])
```

### Default Indicator
After merging the datasets, dropping unknown values, and filtering to Q1 2015, I then create a default flag, defined as loans that are ever 90+ days delinquent in the four-year period from their respective origination dates.

```python
# Track performance of each loan for four years from origination date

four_years = relativedelta(months=+48)

merged_orig_Q1_2015_nonan_stat_fix['Cutoff_Date'] = merged_orig_Q1_2015_nonan_stat_fix['Orig_Date'] + four_years

merged_orig_Q1_2015_nonan_stat_fix_19 = merged_orig_Q1_2015_nonan_stat_fix[merged_orig_Q1_2015_nonan_stat_fix['Date']<=merged_orig_Q1_2015_nonan_stat_fix['Cutoff_Date']]
```

```python
# Create default flag

merged_orig_Q1_2015_nonan_stat_fix_19['Default_Flag'] = np.where(merged_orig_Q1_2015_nonan_stat_fix_19['Curr_Loan_Status']>=3, 1, 0)

'''Loans that ever default'''
'''All other metrics static'''
loan_level_defaults = merged_orig_Q1_2015_nonan_stat_fix_19.groupby('Loan_ID',as_index=False).max()
```

```python
# Drop NaN FICO & DTI values

loan_level_defaults_1 = loan_level_defaults[loan_level_defaults['Orig_FICO'].notna() & loan_level_defaults['Orig_DTI'].notna()]
```

```python
# Functions for chart value labels

def bar_labels_perc(rects):
    for rect in rects:
        height = rect.get_height()
        plt.annotate(
        f'{round(float(height)*100,2)}%',                     
        (rect.get_x() + rect.get_width()/2, height),         
        xytext=(0, 2),         
        textcoords="offset points", 
        ha='center',                
        va='bottom') 
        
def bar_labels_count(rects):
    for rect in rects:
        height = rect.get_height()
        plt.annotate(
        f'{round(int(height),0)}',                      
        (rect.get_x() + rect.get_width()/2, height),         
        xytext=(0, 2),          
        textcoords="offset points", 
        ha='center',                
        va='bottom') 
```

```python
fig,ax = plt.subplots()

tot_loan_ct = total_loans_ins['Loan_Counts'].sum()

unins_perc = 'Uninsured'
rect1 = plt.bar(unins_perc,total_loans_ins['Loan_Counts'].iloc[0]/tot_loan_ct, color='midnightblue')
ins_borr_perc = 'Insured - Borrower Paid'
rect2 = plt.bar(ins_borr_perc,total_loans_ins['Loan_Counts'].iloc[1]/tot_loan_ct, color='tomato')
ins_lend_perc = 'Insured - Lender Paid'
rect3 = plt.bar(ins_lend_perc,total_loans_ins['Loan_Counts'].iloc[2]/tot_loan_ct, color='gold')

bar_labels_perc(rect1)
bar_labels_perc(rect2)
bar_labels_perc(rect3)

ax.set_ylim(0,.9)
ax.set_title("Total Percent of Loans by Insurance Type \n Q1 2015 - Q1 2019")
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
fig.tight_layout()
plt.show()
```

```python
# Function to plot counts and proportions

def bar_plots(first_series,second_series,bartype,column,flabel,slabel,fcolor,scolor,title,tseries=False,third_series=None,tlabel=None,tcolor=None):
    ax,fig = plt.subplots(figsize=(7,5))
    
    if bartype == 'proportion':
        rect1 = plt.bar(flabel,first_series[column].sum()/first_series[column].count(),label=flabel,color=fcolor)
        rect2 = plt.bar(slabel,second_series[column].sum()/second_series[column].count(),label=slabel,color=scolor)
        
        bar_labels_perc(rect1)
        bar_labels_perc(rect2)
        
        if tseries:
            rect3 = plt.bar(tlabel,third_series[column].sum()/third_series[column].count(),label=tlabel,color=tcolor)
            bar_labels_perc(rect3)
        
        ax = plt.gca()
        ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
        ax.set_title(f'Percent of {title}')
 

    elif bartype == 'count':
        rect1 = plt.bar(flabel,first_series[column].sum(),label=flabel,color=fcolor)
        rect2 = plt.bar(slabel,second_series[column].sum(),label=slabel,color=scolor)
        
        bar_labels_count(rect1)
        bar_labels_count(rect2)
        
        if tseries:
            rect3 = plt.bar(tlabel,third_series[column].sum(),label=tlabel,color=tcolor)
            bar_labels_count(rect3)
        
        ax = plt.gca()
        ax.set_title(f'Number of {title}')
    
    plt.tight_layout()
    if tseries:
        plt.savefig(f'images/{flabel} v. {slabel} v. {tlabel}: {title}.png')
    else:
        plt.savefig(f'images/{flabel} v. {slabel}: {title}.png')
    plt.show()
    
# Plot Uninsured v. Insured (Borrower-Paid) v. Insured (Lender-Paid)
bar_plots(loan_level_defaults_unins,loan_level_defaults_ins_borr,'count',
    'Count','Uninsured','Insured - Borrower Paid','midnightblue','tomato',
    'Loans by Insurance Type \n Q1 2015 - Q1 2019',True,loan_level_defaults_ins_lend,
    'Insured - Lender Paid','gold')

bar_plots(loan_level_defaults_unins,loan_level_defaults_ins_borr,'proportion',
    'Default_Flag','Uninsured','Insured - Borrower Paid','midnightblue','tomato',
    'Loan Defaults by Insurance Type \n Q1 2015 - Q1 2019',True,loan_level_defaults_ins_lend,
    'Insured - Lender Paid','gold')

bar_plots(loan_level_defaults_unins,loan_level_defaults_ins_borr,'count',
    'Default_Flag','Uninsured','Insured - Borrower Paid','midnightblue','tomato',
    'Loan Defaults by Insurance Type \n Q1 2015 - Q1 2019',True,loan_level_defaults_ins_lend,
    'Insured - Lender Paid','gold')

# Plot Insured v. Uninsured
bar_plots(loan_level_defaults_unins,loan_level_defaults_ins,'count','Default_Flag','Uninsured','Insured','midnightblue','tomato','Loan Defaults')
bar_plots(loan_level_defaults_unins,loan_level_defaults_ins,'proportion','Default_Flag','Uninsured','Insured','midnightblue','tomato','Loan Defaults')
bar_plots(loan_level_defaults_unins,loan_level_defaults_ins,'proportion','Primary_Flag','Uninsured','Insured','midnightblue','tomato','Primary Residence Properties')
bar_plots(loan_level_defaults_unins,loan_level_defaults_ins,'proportion','First_Time_Home_Flag','Uninsured','Insured','midnightblue','tomato','First-Time Home Buyers')

# Plot Insured (Borrower-Paid Prem) v. Insured (Lender-Paid Prem)
bar_plots(loan_level_defaults_ins_borr,loan_level_defaults_ins_lend,'count','Default_Flag','Insured(Borrower)','Insured(Lender)','yellowgreen','darkgreen','Loan Defaults')
bar_plots(loan_level_defaults_ins_borr,loan_level_defaults_ins_lend,'proportion','Default_Flag','Insured(Borrower)','Insured(Lender','yellowgreen','darkgreen','Loan Defaults')
bar_plots(loan_level_defaults_ins_borr,loan_level_defaults_ins_lend,'proportion','Primary_Flag','Insured(Borrower)','Insured(Lender)','yellowgreen','darkgreen','Primary Residence Properties')
bar_plots(loan_level_defaults_ins_borr,loan_level_defaults_ins_lend,'proportion','First_Time_Home_Flag','Insured(Borrower)','Insured(Lender)','yellowgreen','darkgreen','First-Time Home Buyers')
```

```python
# Function to plot means

def means_plots(column,title,first_series,second_series,flabel,slabel,fcolor,scolor,axlabel,filename,f_xpos,s_xpos,f_ypos,s_ypos,):
    ax,fig = plt.subplots(figsize=(7,5))

    sns.distplot(first_series[column],bins=30,label=flabel,color=fcolor,axlabel=axlabel)
    sns.distplot(second_series[column],bins=30,label=slabel,color=scolor,axlabel=axlabel)

    plt.axvline(first_series[column].mean(), color='k', linestyle='dashed', linewidth=1)
    plt.axvline(second_series[column].mean(), color='k', linestyle='dashed', linewidth=1)

    plt.ylabel("Density")
    min_ylim, max_ylim = plt.ylim()
    plt.text(first_series[column].mean()*f_xpos, max_ylim*f_ypos, 'Mean {} \n({}):\n {:.0f}'.format(axlabel,flabel,first_series[column].mean()))
    plt.text(second_series[column].mean()*s_xpos, max_ylim*s_ypos, 'Mean {} \n({}): \n{:.0f}'.format(axlabel,slabel,second_series[column].mean()))

    ax = plt.gca()
    ax.legend(loc='upper left')
    ax.set_title(f'Distribution of {title}')
    plt.tight_layout()
    plt.savefig(f'images/{filename}')
    plt.show()

# Plot means for uninsured vs. insured loans
means_plots('Orig_FICO','FICO Scores',loan_level_defaults_unins,loan_level_defaults_ins,'Uninsured','Insured','midnightblue','orangered','FICO','Insured v. Uninsured_FICO.png',1.005,0.94,0.87,0.87)    
means_plots('Orig_LTV','Loan-to-Value (LTV) Ratios',loan_level_defaults_unins,loan_level_defaults_ins,'Uninsured','Insured','midnightblue','orangered','LTV','Insured v. Uninsured_LTV.png',0.7,0.8,0.87,0.87)
means_plots('Orig_UPB','Unpaid Principal Balance (UPB) Ratios',loan_level_defaults_unins,loan_level_defaults_ins,'Uninsured','Insured','midnightblue','orangered','UPB','Insured v. Uninsured_UPB.png',1.09,1.13,0.88,0.73)    
means_plots('Orig_DTI','Debt-to-Income (DTI) Ratios',loan_level_defaults_unins,loan_level_defaults_ins,'Uninsured','Insured','midnightblue','orangered','DTI','Insured v. Uninsured_DTI.png',0.75,1.03,0.8,0.8)    

# Plot means for insured (borrower-paid prem) vs. insured (lender-paid prem) loans
means_plots('Orig_FICO','FICO Scores',loan_level_defaults_ins_borr,loan_level_defaults_ins_lend,'Insured \n(Borrower)','Insured \n(Lender)','yellowgreen','darkgreen','FICO','Insured(Borrower) v. Insured(Lender)_FICO.png',0.94,1.01,0.82,0.82)    
means_plots('Orig_LTV','Loan-to-Value (LTV) Ratios',loan_level_defaults_ins_borr,loan_level_defaults_ins_lend,'Insured \n(Borrower)','Insured \n(Lender)','yellowgreen','darkgreen','LTV','Insured(Borrower) v. Insured(Lender)_LTV.png',0.88,1.03,0.7,0.7)
means_plots('Orig_UPB','Unpaid Principal Balance (UPB)',loan_level_defaults_ins_borr,loan_level_defaults_ins_lend,'Insured \n(Borrower)','Insured \n(Lender)','yellowgreen','darkgreen','UPB','Insured(Borrower) v. Insured(Lender)_UPB.png',0.5,1.02,0.7,0.7)
means_plots('Orig_DTI','Debt-to-Income (DTI) Ratios',loan_level_defaults_ins_borr,loan_level_defaults_ins_lend,'Insured \n(Borrower)','Insured \n(Lender)','yellowgreen','darkgreen','DTI','Insured(Borrower) v. Insured(Lender)_DTI.png',0.73,1.05,0.8,0.8)

```

### Hypothesis Testing: Uninsured vs. Insured Loans

```python
'''
Hypothesis Testing: Uninsured vs. Insured
Welch's t-test & Mann-Whitney U Test
'''

'''CREDIT RISK METRICS'''

'''
Default Flag
H0: Uninsured loans default as just often as insured loans.
HA: Uninsured loans are less likely to default than insured loans.
'''
default_pval_t = stats.ttest_ind(loan_level_defaults_unins['Default_Flag'],loan_level_defaults_ins['Default_Flag'],equal_var=False)
print(f'Default t-test pval: {default_pval_t}')
default_pval_m_ = stats.mannwhitneyu( loan_level_defaults_unins['Default_Flag'],loan_level_defaults_ins['Default_Flag'],alternative="two-sided")
print(f'Default Mann-Whitney pval: {default_pval_m_}')

'''
LTV
H0: Uninsured loans have equal or higher LTV scores than insured loans. I.e.
P(Unins Default Rate <= Ins Default Rate)
HA: Uninsured loans have significantly lower LTV scores than insured loans.
'''
ltv_pval_t = stats.ttest_ind(loan_level_defaults_unins['Orig_LTV'],loan_level_defaults_ins['Orig_LTV'],equal_var=False)
print(f'LTV t-test pval: {ltv_pval_t}')
ltv_pval_m_ = stats.mannwhitneyu( loan_level_defaults_unins['Orig_LTV'],loan_level_defaults_ins['Orig_LTV'], alternative="two-sided")
print(f'LTV Mann-Whitney pval: {ltv_pval_m_}')

'''
DTI
H0: Uninsured loans have equal or higher DTI ratios than insured loans. I.e.
P(Unins Default Rate <= Ins Default Rate)
HA: Uninsured loans have lower DTI ratios than insured loans.
'''
dti_pval_t = stats.ttest_ind(loan_level_defaults_unins['Orig_DTI'],loan_level_defaults_ins['Orig_DTI'],equal_var=False)
print(f'DTI t-test pval: {dti_pval_t}')
dti_pval_m_ = stats.mannwhitneyu( loan_level_defaults_unins['Orig_DTI'],loan_level_defaults_ins['Orig_DTI'], alternative="two-sided")
print(f'DTI Mann-Whitney pval: {dti_pval_m_}')

'''
Original Unpaid Principal Balance
H0: Uninsured loans have equal or higher UPBs than insured loans. I.e.
P(Unins Default Rate <= Ins Default Rate)
HA: Uninsured loans have lower UPBs average than insured loans.
'''
upb_pval_t = stats.ttest_ind(loan_level_defaults_unins['Orig_UPB'],loan_level_defaults_ins['Orig_UPB'],equal_var=False)
print(f'UPB t-test pval: {upb_pval_t}')

upb_pval_m_ = stats.mannwhitneyu( loan_level_defaults_unins['Orig_UPB'],loan_level_defaults_ins['Orig_UPB'], alternative="two-sided")
print(f'UPB Mann-Whitney pval: {upb_pval_m_}')

'''
First-time Home Buyer Flag
H0: There are just as many or more first-time home buyers backing uninsured loans than there are backing insured loans. 
HA: There are fewer first-time home buyers backing uninsured loans than there are backing insured loans. 
'''
first_pval_t = stats.ttest_ind(loan_level_defaults_unins['First_Time_Home_Flag'],loan_level_defaults_ins['First_Time_Home_Flag'],equal_var=False)
print(f'First Home t-test: {first_pval_t}')

first_pval_m_ = stats.mannwhitneyu( loan_level_defaults_unins['First_Time_Home_Flag'],loan_level_defaults_ins['First_Time_Home_Flag'], alternative="two-sided")
print(f'First Home t-test: {first_pval_m_}')


'''CREDIT WORTHINESS METRICS'''

'''
FICO
H0: Uninsured loans have equal or lower FICO scores than insured loans. I.e.
P(Unins Default Rate <= Ins Default Rate)
HA: Uninsured loans have higher FICO scores than insured loans.
'''
fico_pval_t = stats.ttest_ind(loan_level_defaults_unins['Orig_FICO'],loan_level_defaults_ins['Orig_FICO'],equal_var=False)
print(f'FICO t-test pval: {fico_pval_t}')
fico_pval_m_ = stats.mannwhitneyu( loan_level_defaults_unins['Orig_FICO'],loan_level_defaults_ins['Orig_FICO'], alternative="two-sided")
print(f'FICO Mann-Whitney pval: {fico_pval_m_}')

'''
Primary Residence Flag
H0: Uninsured loans have equal or fewer primary residences than insured loans. I.e.
P(Unins Default Rate <= Ins Default Rate)
HA: Uninsured loans have higher primary residences than insured loans.
'''
prim_pval_t = stats.ttest_ind(loan_level_defaults_unins['Primary_Flag'],loan_level_defaults_ins['Primary_Flag'],equal_var=False)
print(f'Primary Residence pval: {prim_pval_t}')
prim_pval_m_ = stats.mannwhitneyu( loan_level_defaults_unins['Primary_Flag'],loan_level_defaults_ins['Primary_Flag'], alternative="two-sided")
print(f'Primary Residence pval: {prim_pval_m_}')
```

#### Bonferroni Correction: Insured vs. Uninsured Loans

```python
'''Bonferroni Correction: Insured v. Uninsured'''

print(f'With seven metrics, the probability that we get at least one significant result due to random chance is {round((1-(1-0.05)**7)*100,2)}%!\n')

p_labels = ['default','ltv','fico','dti','upb','primary','first-home']
pvals_t = [default_pval_t[1], ltv_pval_t[1], fico_pval_t[1], dti_pval_t[1], upb_pval_t[1], prim_pval_t[1], first_pval_t[1]]
pvals_m_ = [default_pval_m_[1], ltv_pval_m_[1], fico_pval_m_[1], dti_pval_m_[1], upb_pval_m_[1], prim_pval_m_[1], first_pval_m_[1]]

t_dict = dict()
m_dict = dict()

for lab, pval in zip(p_labels,pvals_t):
    t_dict[lab] = pval

for lab, pval in zip(p_labels,pvals_m_):
    m_dict[lab] = pval

# Adjusted p-values
p_adjusted_t = multipletests(pvals_t, alpha=0.05/7, method='bonferroni')
p_adjusted_m_ = multipletests(pvals_m_, alpha=0.05/7, method='bonferroni')

t_dict_adj = dict()
m_dict_adj = dict()

for lab, pval in zip(p_labels,p_adjusted_t[1]):
    t_dict_adj[lab] = pval

for lab, pval in zip(p_labels,p_adjusted_m_[1]):
    m_dict_adj[lab] = pval
    
print('Original (Uninsured v. Insured):')
print(t_dict)
print(m_dict)
print('\n')
print('Bonferroni Correction (Uninsured v. Insured):')
print(t_dict_adj)
print(m_dict_adj)
    
    
```
